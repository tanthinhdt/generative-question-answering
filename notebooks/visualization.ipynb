# T5 Small

```python
import os
import json
import matplotlib.pyplot as plt

# change these variables before visualization
entries = ['20230502E001', '20230503E001', '20230503E002']
start = 0
eval_steps = 1000
outputs_dir = '../outputs/t5/t5_small'

history = {
    'train_loss': [],
    'eval_loss': [],
    'rougeL': [],
    'rougeLsum': []
}
for entry in entries:
    log_path = os.path.join(outputs_dir, entry, f'{entry}.json')
    with open(log_path, 'r') as f:
        log = json.load(f)
    history['train_loss'].extend(log['train_loss'])
    history['eval_loss'].extend(log['eval_loss'])
    history['rougeL'].extend(log['rougeL'])
    history['rougeLsum'].extend(log['rougeLsum'])

steps = [i * eval_steps + start for i in range(len(history['eval_loss']))]
f, ax = plt.subplots(2, 2, figsize=(20, 12))
ax[0][0].plot(steps, history['train_loss'], label='train_loss')
# ax[0][0].plot(steps, history['eval_loss'], color='orange', label='eval_loss')
ax[0][0].set_title('train_loss')
ax[0][0].set_xlabel('steps')

ax[0][1].plot(steps, history['eval_loss'])
ax[0][1].set_title('eval_loss')
ax[0][1].set_xlabel('steps')

ax[1][0].plot(steps, history['rougeL'])
ax[1][0].set_title('rougeL')
ax[1][0].set_xlabel('steps')

ax[1][1].plot(steps, history['rougeLsum'])
ax[1][1].set_title('rougeLsum')
ax[1][1].set_xlabel('steps')

plt.show()
```
